### Plot samples for models trained on the complete MHD dataset

```python
from multivae.data.datasets import MHD
from config import data_path
```

```python
dataset = MHD(data_path, split='test')
```

```python
from multivae.models import AutoModel
import os, glob
import json
from config import *
from multivae.metrics.reconstruction import Reconstruction, ReconstructionConfig
from multivae.metrics import Visualization, VisualizationConfig
import wandb
import matplotlib.pyplot as plt
import numpy as np

models_name = ['JMVAE', 'JNF','MMVAE','MVAE','MVTCAE','MoPoE','MMVAEPlus']
betas = ['beta_5']
rescales = ['True']

fig = plt.figure(figsize=(1.5*5,5*5))
i=1
for model_name in models_name:
    for beta in betas:
        for rescale in rescales:
  
            for seed in range(1):
                try:
                    model_path = f'asenella/{model_name}_{beta}_scale_{rescale}_seed_{seed}'
                    model = AutoModel.load_from_hf_hub(model_path, allow_pickle=True)
                
                except:
                    model_path = f'asenella/mhd_config_1_{model_name}_{beta}_scale_{rescale}_seed_{seed}'
                    model = AutoModel.load_from_hf_hub(model_path, allow_pickle=True)
                          
                
                if seed==0:
                    vis_config = VisualizationConfig(
                        batch_size=128,
                        n_samples=3,
                        n_data_cond=4
                    )
                    
                    vis_module = Visualization(
                        model,
                        dataset,
                        # output=path,
                        eval_config=vis_config
                    )

                    
                    image = vis_module.conditional_samples_subset(['audio'], ['image', 'audio'])
                    
                    ax = plt.subplot(7,1,i)
                    

                    ax.imshow(np.array(image))
                    
                    ax.set_xticks([])
                    ax.set_yticks([])

                    
                    ax.set_ylabel(model_name)
                    
                    i+=1
                    
plt.tight_layout()

plt.savefig("recon_from_audio")
                
            
            
                
```
